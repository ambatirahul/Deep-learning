```python
plt.figure(figsize=(8,6))
sns.countplot(fer.emotion)
plt.title('Count for different Emotions')
plt.tight_layout()
plt.grid(True)
```

```python
plt.subplots(figsize=(10,8))
emotion_cnt = fer['emotion'].groupby(fer['emotion']).count()
plt.axis('equal')
plt.pie(emotion_cnt, startangle=45, labels=emotionList, autopct='%1.1f%%')
plt.legend(emotionList, loc=4)
plt.title('Share of Emotions')
plt.show()
```

### From the above plots we se there are a lot of examples for happy but not for disgust.

### Data set has sub-sets for training and testing.

```python
fer.pixels = fer.pixels.apply(lambda i: np.fromstring(i, sep=' '))
fer.head()
```

```python
pixels = np.vstack(fer.pixels.values)
pixels = pixels.reshape(-1,48,48,1)
emotions = np.array(fer.emotion)
print(pixels.shape)
```

```python
plt.figure(0, figsize=(12,8))
for i in range(1,17):
    plt.subplot(4,4,i)
    plt.imshow(pixels[i, :, :, 0], cmap='gray')
    plt.title(emotion[emotions[i]])
plt.tight_layout()
plt.show()
```

### Above images are a sample set of the data and we can visually make out their expressions.

We looked at some sample images and we can answer those expressions by looking at the image, we need to build a model to do the same. Predicting expressions on it's own.
